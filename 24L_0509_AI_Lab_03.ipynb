```python
# Question 1- a) Converting the list into a Numpy array and displaying it's shape and the number of dimensions\
import numpy as np 
a = np.array([[12,7,5], [4,15,18], [10,3,14]])
print("The Shape of the Array is = ", a.shape) 
print("The dimenions of the Array are = ", a.ndim)

```

```python
# b) Replacing all the values greater than 10 with 0 and displaying the updated array 
a[a>10] = 0 
print("The updated array after replacing all the values is ", a)
```

```python
# Question 2 - Performing the element wise addition and multiplication of X and Y 
X = np.array([[1,4], [9,16]])
Y = np.array([[2,3], [5,7]])

# Performing element wise addition and multiplication of X and Y 
print(X + Y)   
print(np.add(X,Y))

#Computing the square root of array X 
print(np.sqrt(X))
```

```python
#Question 3 - Pandas
marks = { 
       "Ali": 78, 
       "Sara": 92,
       "Ahmed": 67, 
       "Hina": 85, 
       "Usman": 74
}

# a) Converting the dictionary into a pandas series and displaying the students scoring above 75 

#converting the dictionary to a pandas series 
score_series = pd.Series(marks)

#Filter for the scores grerater than 75 
high_scorers = score_series[score_series >75]

print("Scores above 75")
print(high_scorers)

#b) Caluclating the mean and the maximum marks here 
mean_marks = score_series.mean()
max_marks = score_series.max()

print(f"The mean marks are, {mean_marks}")
print(f"The maximum marks are, {max_marks}")
```

```python
#Question 4 - 
data = { 
    "Name": ["Ayesha", "Bilal", "Kiran", "Omar", "Zara"], 
    "Age": [25,30,27,35,29], 
    "Department": ["HR", "IT", "Finance", "IT", "Marketing"], 
    "Salary": [50000, 70000, 60000, 80000, 65000]
}

# a. Create a DataFrame and display employees from the IT department 
df = pd.DataFrame(data, columns=["Name", "Age", "Department", "Salary"])
#print(df.info())
#print(df)

#displaying employees from the IT department 
it_employees = df[df['Department']== 'IT']
print(it_employees)


#b. Finding the average salary and and sorting the DataFrame in Descending order 
average = df['Salary'].mean()
print(average)

print("Sorted DataFrame in the order of decreasing salary")
df.sort_values(by='Salary', ascending = False )
```
